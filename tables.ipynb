```python
from tables import clean_table, add_diffs

rows = ['Male', 'Female']
columns = ['Aged 12-17 (2023)', 'Aged 12-17 (2024)', 
           'Aged 18+ (2023)', 'Aged 18+ (2024)']

def compile_tables(tables):
    table_dict = {}

    for table in tables:
        number = f'1.{table}B'
        filename = f'table_{number}.csv'
        title = filename_to_title[filename]
        table_info = parse_table_title(title)
        desc = table_info['full_description']
        key = desc.split(" in ")[0]
        
        df = pd.read_csv(f"data/{filename}", index_col=0)
        df = clean_table(df)
        df = df.loc[rows, columns]
        df = add_diffs(df)
        table_dict[key] = df
        print(table_info['time_period'], key)

    return pd.concat(table_dict)
```

```python
tables = range(23, 42, 3)
compile_tables(tables)
```
